## Working with FastAI on CIFAR100 dataset
A dataset of 32x32 rgb images with 100 classes. https://www.cs.toronto.edu/~kriz/cifar.html

```python
from fastai import *
from fastai.vision import *

path = untar_data(URLs.CIFAR_100)
data = ImageDataBunch.from_folder(path=path, train='train', valid='test', size=32, bs=512).normalize(cifar_stats)
```

```python
learn = cnn_learner(data, models.resnet101, metrics=accuracy).to_fp16()
```

```python
learn.freeze()
learn.fit_one_cycle(5)
learn.unfreeze()
learn.fit_one_cycle(5, max_lr = slice(1e-4, 1e-3))
```

```python
data = ImageDataBunch.from_folder(path=path, train='train', valid='test', size=64, bs=512).normalize(cifar_stats)
learn.data = data
learn.to_fp16()
```

```python
learn.freeze()
learn.fit_one_cycle(5)
learn.unfreeze()
learn.fit_one_cycle(5, max_lr = slice(1e-4, 1e-3))
```
